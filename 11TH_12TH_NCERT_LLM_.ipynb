```python
import fitz
import nltk
nltk.download('punkt')  # Download the necessary resources
from nltk.tokenize import sent_tokenize
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import numpy as np
import pandas as pd
import chromadb
import google.generativeai as genai
import os
from IPython.display import Markdown
import torch
import streamlit as st
```

```python
pdf_files = ['el_c.pdf', 'tw_c_nob.pdf','tbio.pdf']
data = ""

# Function to extract text from a single PDF
def extract_text_from_pdf(pdf_path):
    text=""
    with fitz.open(pdf_path) as pdf_document:
        for page_num in range(len(pdf_document)):
            page = pdf_document.load_page(page_num)
            text += page.get_text()
    return text

# Loop through each uploaded PDF file and extract the text
for pdf_file in pdf_files:
    text = extract_text_from_pdf(pdf_file)
    print(f"Length of text from {pdf_file}: {len(text)}")
    data += text
```

```python
print(len(data))
```

```python
# Replace '\n' with ' ' in the data variable
data = data.replace('\n', ' ').strip()

# Optionally, you can print the first 1000 characters to check the result
print(data[-200:])
```

```python
# Tokenize the text into sentences
sentences = sent_tokenize(data)

# Convert sentences to strings if needed
sentences = [str(sentence) for sentence in sentences]
```

```python
print(len(sentences))
```

```python
sentence_chunks=[]
chunk_size=200

for i in range(0,len(sentences),chunk_size):
    chunk_dict={}
    combined_sent=' '.join(sentences[i:i+chunk_size])
    chunk_dict["chunk_token_count"]=len(combined_sent)/4
    chunk_dict["sentence"]=combined_sent
    sentence_chunks.append(chunk_dict)

for i in sentence_chunks[0:3]:
    print(i)
```

```python
print(len(sentence_chunks))
```

```python
embedding_model=SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
for item in tqdm(sentence_chunks):
    item['embedding']=embedding_model.encode(item['sentence'])
text_chunks_and_embeddings_df=pd.DataFrame(sentence_chunks)
text_chunks_and_embeddings_df.head()
```

```python
# Set up a persistent client by specifying a directory for storage
client = chromadb.PersistentClient(path="path/chromadb_ncert")

# Create or connect to a collection
collection = client.create_collection("ncert_collection",metadata={"hnsw:space": "cosine"},)

# Iterate over the dataframe rows and prepare data for ChromaDB
for index, row in text_chunks_and_embeddings_df.iterrows():
    metadata = {
        "sentence": row['sentence'],
    }

    # Add the embedding, with an ID and metadata
    collection.add(
        embeddings=[row['embedding'].tolist()],
        ids=[str(index)],
        metadatas=[metadata]            # Metadata dictionary
    )

print("Data added to ChromaDB with metadata.")
```

```python
with open('app.py', 'w') as f:
    f.write("""
import streamlit as st
import os
import google.generativeai as genai
import chromadb

from sentence_transformers import SentenceTransformer
embedding_model=SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Reconnect to the persistent client
client = chromadb.PersistentClient(path="path/chromadb_ncert")

# Connect to the collection
collection = client.get_collection("ncert_collection")


def query_prompt(question):
    query_embedding = embedding_model.encode(question).tolist()

    results = collection.query(
    query_embeddings=query_embedding,
    n_results=3
    )

    context = ""

    for metadata in results['metadatas'][0]:
        context += metadata['sentence'] + " "

    context = context.strip()

    #context=''
    #for i in results['ids'][0]:
        #context+=text_chunks_and_embeddings_df.iloc[int(i)]['sentence']

    prompt=f'''You are a NEET Paper Setter,
    write an answer for this question:{question}?
    use clear and proper advices from the following context:{context}
    Only answer what is asked.
    Answer everything but strictly specify if its from the context provided to you or no.
    Say 'Thanks' at the end
    '''
    return prompt

# Initialize Gemini-Pro
genai.configure(api_key="YOUR_API_KEY")
model = genai.GenerativeModel('gemini-1.5-flash')

# Gemini uses 'model' for assistant; Streamlit uses 'assistant'
def role_to_streamlit(role):
    return "assistant" if role == "model" else role


# Add a Gemini Chat history object to Streamlit session state
if "chat" not in st.session_state:
    st.session_state.chat = model.start_chat(history=[])

# Display Form Title
st.title("NCERT chatbot")

# Display chat messages from history above current input box
for message in st.session_state.chat.history:
    role = role_to_streamlit(message["role"])
    with st.chat_message(role):
        st.markdown(message["parts"][0]["text"])

# Accept user's next message
if user_input := st.chat_input("Ask anything from NCERT?"):
    # Display user's message
    st.chat_message("user").markdown(user_input)

    # Generate the prompt and get the response from the model
    prompt = query_prompt(user_input)
    response = model.generate_content(prompt)

    # Extract the response text safely
    response_text = ""
    if response and response.candidates and response.candidates[0].content.parts:
        response_text = response.candidates[0].content.parts[0].text
    else:
        response_text = "Sorry, I couldn't generate a response. Please try again."

    # Add the question and response to chat history
    st.session_state.chat.history.append(
        {"role": "user", "parts": [{"text": user_input}]}
    )
    st.session_state.chat.history.append(
        {"role": "model", "parts": [{"text": response_text}]}
    )

    # Display the assistant's response
    with st.chat_message("assistant"):
        st.markdown(response_text)
    """)
```
